```python
import pandas as pd
```

```python
df = pd.read_csv("Points.csv", sep= ';')
df.index = df['idNode']
```

```python
df_new = pd.DataFrame(columns = df.columns)
```

```python
df
```

```python
def getNodes(route, df_new, df):
  actual = route[0]
  i = 1
  while actual != route[-1]:
    if ';' in df.loc[[actual]][['next']].values[0][0]:
      actual = route[i]
      i += 1

    else:
      actual = int(df.loc[[actual]][['next']].values[0][0])
    df_new.loc[actual] = df.loc[actual]
  return(df_new)
    
```

```python
df_new = getNodes(route1_k,df_new,df)
df_new = getNodes(route1_s,df_new,df)
df_new = getNodes(route2_c,df_new,df)
df_new = getNodes(route2_s,df_new,df)
df_new = getNodes(route3_b,df_new,df)
df_new = getNodes(route3_k,df_new,df)
```
